```python
# imports
import os
import requests
from IPython.display import JSON
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
from sqlite3 import Error
```

```python
# retrieve POI data
yelp_df = pd.read_csv('../data/yelp.csv')
foursquare_df = pd.read_csv('../data/foursquare.csv')
```

Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
# remove duplicates from dataframes
# sort by distance to nearest bike station
yelp_df_sort = yelp_df.sort_values('distance')
foursquare_df_sort = foursquare_df.sort_values('distance')
# drop duplicates that are farther from a bike station
drop_cols = ['name', 'address', 'rating', 'price', 'lat_lon']
yelp_df_unique = yelp_df_sort.drop_duplicates(subset = drop_cols, keep = 'first')
foursquare_df_unique = foursquare_df_sort.drop_duplicates(subset = drop_cols, keep = 'first')
```

```python
# standardize lat_lon column
# lat_lon significant digits info https://gis.stackexchange.com/questions/8650/measuring-accuracy-of-latitude-and-longitude
# find least significant digits available
yelp_ll = yelp_df_unique['lat_lon'].values
foursquare_ll = foursquare_df_unique['lat_lon'].values
all_ll = np.append(yelp_ll, foursquare_ll)
least_sigdigs = 99
for ll in all_ll:
    lat, lon = ll.split(',')
    lat_int, lat_dec = lat.split('.')
    lon_int, lon_dec = lon.split('.')
    least_sigdigs = min(least_sigdigs, len(lat_dec), len(lon_dec))
# round all lat_lon coordinates to the least significant digits
def round_lat_lon(ll, sigdigs):
    lat, lon = ll.split(',')
    rounded_lat = round(float(lat), sigdigs)
    rounded_lon = round(float(lon), sigdigs)
    rounded_ll = str(rounded_lat) + ',' + str(rounded_lon)
    return rounded_ll
yelp_df_unique['lat_lon'] = yelp_df_unique['lat_lon'].apply(lambda ll: round_lat_lon(ll, least_sigdigs))
foursquare_df_unique['lat_lon'] = foursquare_df_unique['lat_lon'].apply(lambda ll: round_lat_lon(ll, least_sigdigs))
```

```python
# join dataframes based on lat_lon as key
joined_df = pd.merge(foursquare_df_unique, yelp_df_unique, on=['lat_lon', 'address', 'station_idx'], how='outer', suffixes=['_f', '_y'])
joined_df
```

```python
# found no name overlap, combine name columns
name_df = None
if joined_df[joined_df['name_f'].notnull() == joined_df['name_y'].notnull()].empty:
    joined_df['name_f'].fillna('', inplace=True)
    joined_df['name_y'].fillna('', inplace=True)
    joined_df['name'] = joined_df['name_f'] + joined_df['name_y']
    name_df = joined_df.drop(columns=['name_f','name_y'])
name_df
```

```python
# Combine duplicates, arbitrarily keep Foursquare's lat_lon
name_df.loc[30,['rating_y', 'price_y', 'review_count', 'distance_y']] = [4.0, 1.0, 19.0, 45.249143]
name_df.loc[26,['rating_y', 'price_y', 'review_count', 'distance_y']] = [4.5, 3.0, 26.0, 54.105705]
name_df.loc[35,['rating_y', 'price_y', 'review_count', 'distance_y']] = [4.0, 3.0, 41.0, 73.482881]
name_df.loc[171,['rating_y', 'price_y', 'review_count', 'distance_y']] = [4.0, 1.0, 9.0, 711.777091]
name_df.drop([186, 188, 190, 264], inplace=True)
```

```python
# combine ratings
name_df['rating_y'] = 2 * name_df['rating_y']
name_df['rating'] = name_df[['rating_f','rating_y']].mean(axis=1, skipna=True)
rating_df = name_df.drop(columns=['rating_f', 'rating_y'])
rating_df
```

```python
# combine price
rating_df['price'] = rating_df[['price_f', 'price_y']].mean(axis=1, skipna=True)
price_df = rating_df.drop(columns = ['price_f', 'price_y'])
price_df
```

```python
# combine distances
price_df['distance'] = price_df[['distance_f', 'distance_y']].mean(axis=1, skipna=True)
final_df = price_df.drop(columns = ['distance_f', 'distance_y'])
final_df
```

```python
# impute popularity
avg_pop = final_df['popularity'].mean()
final_df['popularity'].fillna(avg_pop, inplace=True)
final_df
```

```python
# impute review count
avg_rev = final_df['review_count'].mean()
final_df['review_count'].fillna(avg_rev, inplace=True)
final_df
```

```python
# impute verified
final_df['verified'].fillna(False, inplace=True)
final_df
```

```python
# impute rating
final_df['rating'].replace(-1,np.NaN, inplace=True)
avg_rat = final_df['rating'].mean()
final_df['rating'].fillna(avg_rat, inplace=True)
final_df
```

```python
# impute price 
final_df['price'].replace(-1,np.NaN, inplace=True)
avg_rat = final_df['price'].mean()
final_df['price'].fillna(avg_rat, inplace=True)
final_df
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
# taken from compass, SQL in Python walkthrough
def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")

def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
# connect to database
connection = create_connection("../data/bikes_db.sqlite")
```

```python
# create table for bike stations and insert bike stations csv into stations table
bikes_df = pd.read_csv('../data/citybikes.csv')
bikes_df.to_sql('stations', connection, if_exists='replace', index=False)
```

```python
# create table for places and insert dataframe into table
final_df.to_sql('places', connection, if_exists='replace', index=False)
```

Look at the data before and after the join to validate your data.

```python
# select data to verify
select_stations = "SELECT * FROM stations LIMIT 10"
stations = execute_read_query(connection, select_stations)

for station in stations:
    print(station)
```

```python
select_pois = "SELECT * FROM places LIMIT 10"
pois = execute_read_query(connection, select_pois)

for poi in pois:
    print(poi)
```
